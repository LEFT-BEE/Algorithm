## 비슷한 영화 찾기

```python
str_to_set = lambda s: set(item['name'] for item in eval(s))  

#metadata읽기
meta_data = pd.read_csv("/kaggle/input/the-movies-dataset/movies_metadata.csv")

meta_data['genres'] = meta_data['genres'].apply(str_to_set) #장르를 보기쉽게 set으로 바꿈

meta = meta_data[['id','original_title' , 'original_language' , 'genres']] #->Series라는 타입 = 한줄짜리 데이터프레임
meta = meta.rename(columns = {'id':'movieId' , 'original_title' : 'title' , 'original_language' : 'language'}) #메타 자체를 바꾸는 것이 아니라 조회하는 것 

meta = meta.loc[meta['language'] == 'en',:] #영어만 남김 
#meta.loc[] 특정위치에 데이터를 뽑아낸다 ex)
# meta.loc[3,'title'] 3행 title열 의 데이터 출력 
#meta.loc[3,:] -> 3번행의 모든 열을 가져온다
# meta['title'] = meta.loc[:,'title']
meta['movieId'] = pd.to_numeric(meta['movieId'])

meta
```

```python
keywords = pd.read_csv('/kaggle/input/the-movies-dataset/keywords.csv')
keywords = keywords.rename(columns = {'id' : 'movieId'})

keywords.loc[0,'keywords']

            
keywords['keywords'] = keywords['keywords'].apply(str_to_set)
keywords
```

```python
merged = pd.merge(meta , keywords , on = 'movieId' , how = 'inner')
# inner join , left join , right join , left join등등이 존재
```

```python
dk = merged.loc[merged['title'] == 'The Dark Knight' , :].iloc[0,:]
dkr = merged.loc[merged['title'] == 'The Dark Knight Rises' , :].iloc[0,:]

dk_set = dk.keywords|dk.genres
dkr_set = dk.keywords|dk.genres


def jaccard_similarity(set1 , set2):
    if len(set1 | set2) == 0 : return 0
    return len(set1 & set2) / len(set1 | set2)

print(jaccard_similarity(dk_set , dkr_set))
```
